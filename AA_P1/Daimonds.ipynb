### Step4: Data Description

Based on the basic exploration above, we can define each variables in this dataset.
  
- Price: The price of the Diamond
- Carat: The carat value of the Diamond
- Cut: The cut type of the Diamond, it determines the shine (Poor, Fair, Good, very Good, Excellent)
- Color: The color value of the Diamond (D, E, F, … Z)
- Clarity: The clarity of the Diamond (FL, IF, VVS1, VVS2, VS1, VS2, SI1, SI2, SI3, I1, I2, I3)
- Depth: The depth value of the Diamond

- Table: Flat facet on its surface — the large, flat surface facet that you can see when you look at the diamond from above.

- x: Width of the diamond
- y: Length of the diamond
- z: Height of the diamond

### Step5: Visulaisation Data:
    We will visualize both the distribution of categorical and numerical features in diamond dataset.



#### A) | Heat Map Plot |

```python

plt.figure(figsize=(15,10))
sns.heatmap(Diam.corr(), linecolor="white", linewidths=3,annot=True)
```

#### B) | Scatter Plot |

```python
sns.relplot(x='carat',y='price',hue='clarity',alpha=.7,data=Diam)
plt.xlabel('Weight')
plt.ylabel('Price')
plt.title('Carat Vs Price based on Clarity')
plt.figure(figsize=(30, 10))
```

```python
plt.figure(figsize=(30, 10))
sns.relplot(x='carat',y="price",hue="cut", size = "depth", alpha=.7,data=Diam)
plt.xlabel('Carat Weight')
plt.ylabel('Price')
plt.title('Carat Vs Price based on Cut')
```

#### C) | Bar Plot |

```python
fig, ba_axis = plt.subplots(1, 3, figsize=(15,5))
sns.barplot(x="cut", y="price", data=Diam, ax = ba_axis[0], color='orange')
sns.barplot(x="color", y="price", data=Diam, ax = ba_axis[1], color= 'lightblue')
sns.barplot(x="clarity", y="price", data=Diam, ax = ba_axis[2], color='purple')
```

#### D) | Point Plot |

```python
sns.pointplot(x="cut", y="carat", hue="color", data=Diam, dodge=True)
```

```python
sns.pointplot(x="cut", y="carat", hue="clarity", data=Diam, dodge=True)
```

```python
sns.pointplot(x="color", y="carat", hue="cut", data=Diam, join=False)
```

```python
sns.pointplot(x="color", y="carat", hue="clarity", data=Diam, join=False)
```

#### E) | Histogram Plot |

```python
#plt.figure(figsize=(20,10))
plt.figure(figsize=(10,5))
sns.histplot(x='depth',hue='color',multiple="stack",edgecolor=".5",linewidth=.6,
    log_scale=False,data=Diam)
plt.xlim(55,70)
plt.xlabel('Depth')
plt.ylabel('count')
plt.title('Depth Vs Count')
```
